```python
#As we can see Pike are typically bigger than other species, and the Smelt are samller.
#I think it is make sense to combine the columns Height and Width into a column name Size.
#And after that, lets visualize it agian
df['Size'] = df['Height'] * df['Width']
```

```python
Bream = df[df['Species']=='Bream']
Roach = df[df['Species']=='Roach']
Whitefish = df[df['Species']=='Whitefish']
Parkki = df[df['Species']=='Parkki']
Perch = df[df['Species']=='Perch']
Pike = df[df['Species']=='Pike']
Smelt = df[df['Species']=='Smelt']

plt.figure(figsize=(18,18))
plt.subplot(3,1,3)
plt.scatter(Bream['Size'],Bream['Weight'],c='b',label='Bream')
plt.scatter(Roach['Size'],Roach['Weight'],c='r',label='Roach')
plt.scatter(Whitefish['Size'],Whitefish['Weight'],c='g',label='Whitefish')
plt.scatter(Parkki['Size'],Parkki['Weight'],c='c',label='Parkki')
plt.scatter(Perch['Size'],Perch['Weight'],c='m',label='Perch')
plt.scatter(Pike['Size'],Pike['Weight'],c='y',label='Pike')
plt.scatter(Smelt['Size'],Smelt['Weight'],c='k',label='Smelt')
plt.xlabel('Size', fontsize=12)
plt.ylabel('Weight', fontsize=12)
plt.title('Fish weight based on Size', fontsize=21)
plt.legend(loc='upper left')

plt.tight_layout()
plt.show()
```

```python
#Now drop the columns height and width
df1 = df.drop(['Height','Width'],axis=1)
```

```python
#Now lets see how the weight control the fish species
specise_df = df1.groupby('Species')['Weight'].agg('mean')
specise_df_sorted = specise_df.sort_values(ascending=False)
specise_df_sorted.plot.bar()
plt.ylabel('Mean weight')
plt.title('Mean weight of every species fish',fontsize=21)
plt.show()
#Just like my assumption beforr, the smelt have the lowest stat, and the 
#pike have the highest stat.
```

```python
#Now I am gonna standardize the datapoint
from sklearn.preprocessing import StandardScaler
stdscale = StandardScaler()

df1['scaled_Weight'] = stdscale.fit_transform(df['Weight'].values.reshape(-1,1))
df1['scaled_Length1'] = stdscale.fit_transform(df['Length1'].values.reshape(-1,1))
df1['scaled_Length2'] = stdscale.fit_transform(df['Length2'].values.reshape(-1,1))
df1['scaled_Length3'] = stdscale.fit_transform(df['Length3'].values.reshape(-1,1))
df1['scaled_Size'] = stdscale.fit_transform(df['Size'].values.reshape(-1,1))

df2 = df1.drop(['Weight','Length1','Length2','Length3','Size'],axis=1)
df2.head()
```

```python
x = df2.drop('Species',axis=1)
y = df2['Species']

#Dimensionality reduction
from sklearn.decomposition import PCA
x_dr_pca = PCA(n_components=3, random_state=42).fit_transform(x.values)
```

```python
x_dr_pca = pd.DataFrame(x_dr_pca)
x_dr_pca.head()
```

```python
#Over sampling
from imblearn.over_sampling import RandomOverSampler
balancer = RandomOverSampler()
x_res,y_res = balancer.fit_sample(x_dr_pca,y)
x_res.shape,y_res.shape
```

```python
#Now the classes are balanced
y_res_df = pd.DataFrame(y_res)
y_res_df.value_counts()
```

```python
#Now split the data and train the model
from sklearn.model_selection import StratifiedKFold
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB

spliter = StratifiedKFold(n_splits=5,shuffle=False,random_state=None)
for train_index, test_index in spliter.split(x_res, y_res):
    x_train, x_test = x_res.iloc[train_index], x_res.iloc[test_index]
    y_train, y_test = y_res.iloc[train_index], y_res.iloc[test_index]

x_train = x_train.values
x_test = x_test.values
y_train = y_train.values
y_test = y_test.values

train_unique_label, train_counts_label = np.unique(y_train, return_counts=True)
test_unique_label, test_counts_label = np.unique(y_test, return_counts=True)

print('Label Distributions: \n')
print(train_counts_label/ len(y_train))
print(test_counts_label/ len(y_test))
```

```python
#Hypertuning time
from sklearn.model_selection import GridSearchCV

rfc = GridSearchCV(RandomForestClassifier(), {
    'n_estimators': [100,150,200,250],
    'criterion': ['gini','entropy'],
    'max_depth': [5,10,15]
}, cv=5, return_train_score=False)

svc = GridSearchCV(SVC(), {
    'C': [1,5,10,15],
    'kernel': ['poly', 'rbf', 'sigmoid'],
    'degree': [2,3],
    'gamma': ['scale','auto'],
}, cv=5, return_train_score=False)

nbc = GaussianNB()
```

```python
rfc.fit(x_train,y_train)
svc.fit(x_train,y_train)
nbc.fit(x_train,y_train)
```

```python
print("Random forest classifier best param: {}".format(rfc.best_params_))
print("Support vector classifier best param: {}".format(svc.best_params_))
print("Naive Bayes classifier did not do any tuning")
```

```python
#Now i gonna check the score of each classifier
print("Random forest classifier best score: {}".format(rfc.best_score_))
print("Support vector classifier best score: {}".format(svc.best_score_))
nbc_score = nbc.score(x_test,y_test)
print("Naive Bayes classifier best score: {}".format(nbc_score))
```

```python
from sklearn.metrics import precision_score, recall_score, f1_score

y_pred_rfc = rfc.predict(x_test)
rfc_precision = precision_score(y_test,y_pred_rfc,average='micro')
rfc_recall = recall_score(y_test,y_pred_rfc,average='micro')
rfc_f1 = f1_score(y_test,y_pred_rfc,average='micro')
print("Random forest classifier precision score: {}%".format(round(rfc_precision,2) * 100))
print("Random forest classifier recall score: {}%".format(round(rfc_recall,2) * 100))
print("Random forest classifier f1 score: {}%".format(round(rfc_f1,2) * 100))
print('\n')

y_pred_svc = svc.predict(x_test)
svc_precision = precision_score(y_test,y_pred_svc,average='micro')
svc_recall = recall_score(y_test,y_pred_svc,average='micro')
svc_f1 = f1_score(y_test,y_pred_svc,average='micro')
print("Support vector classifier precision score: {}%".format(round(svc_precision,2) * 100))
print("Support vector classifier recall score: {}%".format(round(svc_recall,2) * 100))
print("Support vector classifier f1 score: {}%".format(round(svc_f1,2) * 100))
print('\n')

y_pred_nbc = nbc.predict(x_test)
nbc_precision = precision_score(y_test,y_pred_nbc,average='micro')
nbc_recall = recall_score(y_test,y_pred_nbc,average='micro')
nbc_f1 = f1_score(y_test,y_pred_nbc,average='micro')
print("Naive Bayes classifier precision score: {}%".format(round(nbc_precision,2) * 100))
print("Naive Bayes classifier recall score: {}%".format(round(nbc_recall,2) * 100))
print("Naive Bayes classifier f1 score: {}%".format(round(nbc_f1,2) * 100))
```

```python
#Random forest work the best and have the average score between 80%-90%
#Now save the model
import joblib
joblib.dump(rfc,'Fish species classifier.pkl')
```
